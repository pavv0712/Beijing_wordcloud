```python
from wordcloud import WordCloud
import jieba
import jieba.posseg as pag
from collections import Counter
```

```python
#크롤링한 데이터 불러오기

lists = ''
with open("C:/Users/나/Desktop/Culture_news.csv","r", encoding='utf-8') as file:
    for i in file.readlines():
        text = i.strip().replace('content', '').replace(',title', '').replace(',(', '').replace(',)', '').replace(',"', '').replace(",''",'').replace('?', '').replace('!', '')
        lists += text
file.close()
```

```python
#jieba 사용하여 형태소 분리하고 합치기

cut = jieba.cut(lists)

text = ','.join(cut)
```

```python
#품사별로 단어 나누기

fenkai = [(x.word, x.flag) for x in pag.cut(text)]
```

```python
fenkai
```

```python
#의미있는 명사, 동명사, 약어만 분류하여 정리하기

n_fenkai = [(x.word,x.flag) for x in pag.cut(text) if x.flag.startswith('n') or x.flag.startswith('vn') or x.flag.startswith('j')]
```

```python
#정리하기

final_text = []
for i in range(len(n_fenkai)):
    final_text.append(n_fenkai[i][0])
    
final_text
```

```python
#단어별 빈도수 구하기

count = Counter(final_text)
words = dict(count.most_common())
words
```

```python
#워드클라우드 생성

cloud = WordCloud(
    width = 3000,
    height = 1500,
    background_color = 'white',
    font_path='C:/Windows/Fonts/simsun.ttc',
    
)

wc = cloud.generate_from_frequencies(words)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
